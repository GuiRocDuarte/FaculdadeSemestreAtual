```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv('C:\\01-FaculdadeSemestreAtual\\Amostragem_Aplicada\\Municipio_Enem_2019_PIB_2019 - Municipio_Enem_2019_PIB_2019.csv')

# Exibir as primeiras linhas para ver a estrutura dos dados
df.head()
```

```python
df.info()
```

```python
# Transformando a coluna em float e removendo as virgulas
df['VL_ADICIONADO_BRUTO_TOTAL'] = df['VL_ADICIONADO_BRUTO_TOTAL'].str.replace(',', '').astype(float)
```

```python
# Visualizar estatísticas básicas da população
pop_mean = df['VL_ADICIONADO_BRUTO_TOTAL'].mean()
pop_std = df['VL_ADICIONADO_BRUTO_TOTAL'].std()

print(f'Média da população: {pop_mean}')
print(f'Desvio padrão da população: {pop_std}')
```

# Aleatória Simples

```python
# Tamanho da amostra
sample_size = 100  # Tamanho da amostra

# Amostragem Aleatória Simples (AAS)
amostra = df.sample(n=sample_size)

# Visualizar estatísticas da amostra
as_sample_mean = amostra['VL_ADICIONADO_BRUTO_TOTAL'].mean()
as_sample_std = amostra['VL_ADICIONADO_BRUTO_TOTAL'].std()

print(f'Média da amostra: {as_sample_mean}')
print(f'Desvio padrão da amostra: {as_sample_std}')
```

```python
# Comparação estatística
print(f'Diferença entre média da população e da amostra: {abs(pop_mean - as_sample_mean)}')
print(f'Diferença entre desvio padrão da população e da amostra: {abs(pop_std - as_sample_std)}')
```

# Estratificada

```python
stratification_column = 'UF'  

# Verificar as categorias na coluna de estratificação
strata = df[stratification_column].unique()
print(f'Estratos encontrados: {strata}')

# Tamanho da amostra total
total_sample_size  = 100  
strata_sample_size = {}

# Calcular o tamanho da amostra para cada estrato, garantindo que não ultrapasse a população disponível
for stratum in strata:
    stratum_data = df[df[stratification_column] == stratum]
    strata_sample_size[stratum] = min(len(stratum_data), int(total_sample_size / len(strata)))

# Amostragem Estratificada
strata_samples = []
for stratum, size in strata_sample_size.items():
    stratum_data = df[df[stratification_column] == stratum]
    if size > 0:  # Verifica se o tamanho da amostra é maior que 0
        stratum_sample = stratum_data.sample(n=size, random_state=42)
        strata_samples.append(stratum_sample)

# Concatenar as amostras de cada estrato em um único DataFrame
amostra = pd.concat(strata_samples)
```

```python
# Visualizar estatísticas da amostra
e_sample_mean = amostra['VL_ADICIONADO_BRUTO_TOTAL'].mean()
e_sample_std = amostra['VL_ADICIONADO_BRUTO_TOTAL'].std()


print(f'Média da amostra: {e_sample_mean}')
print(f'Desvio padrão da amostra: {e_sample_std}')
```

```python
# Comparação estatística
print(f'Diferença entre média da população e da amostra: {abs(pop_mean - e_sample_mean)}')
print(f'Diferença entre desvio padrão da população e da amostra: {abs(pop_std - e_sample_std)}')
```
